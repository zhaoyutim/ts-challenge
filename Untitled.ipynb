```python
from main import load_timeseries_data, drop_zero_columns

df = load_timeseries_data(stock_symbol=["002521"])
df = drop_zero_columns(df)
```

```python
df.head(5)
```

```python
import datetime
df_filtered =df[(df['InstrumentID']==2521) & (df['TradingDay']==datetime.datetime(2022, 6, 1))]
```

```python

import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import dates as mdates
import pandas as pd
import seaborn as sns

df_filtered = df_filtered.sort_values('DataTime').reset_index(drop=True)

all_asks_price = [row[f'AskPrice{n}'] for _, row in df_filtered.iterrows() for n in range(1, 11)]
all_asks_quantity = [row[f'AskVolume{n}'] for _, row in df_filtered.iterrows() for n in range(1, 11)]
all_bids_price = [row[f'BidPrice{n}'] for _, row in df_filtered.iterrows() for n in range(1, 11)]
all_bids_quantity = [row[f'BidVolume{n}'] for _, row in df_filtered.iterrows() for n in range(1, 11)]

artists=[]
fig, ax = plt.subplots()
for i in range(10):
    ask_df = pd.DataFrame({'price': all_asks_price[:10], 'quantity': all_asks_quantity[:10]})
    bid_df = pd.DataFrame({'price': all_bids_price[:10], 'quantity': all_bids_quantity[:10]})
    # Create figure and axis
    ax.set_title(f"Order Book Depth Chart {i}")
    sns.ecdfplot(x="price", weights="quantity", stat="count",
                data=ask_df, ax=ax, color="red")
    sns.ecdfplot(x="price", weights="quantity", stat="count",
                    complementary=True, data=bid_df, ax=ax, color="green")
    # complementary=True allows reflects that lower bids are "better"
    ax.set_xlabel("Price")
    ax.set_ylabel("Quantity")
    artists.append(ax)

ani = animation.ArtistAnimation(fig=fig, artists=artists, interval=1000)
plt.show()
```
