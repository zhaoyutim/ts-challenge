```python
from create_dataset import load_timeseries_data, drop_zero_columns, remove_lunch_break_data
from feature_engineering import create_features
import datetime
import pandas as pd
import numpy as np
import ipywidgets as widgets
from ipywidgets import interact

stock_symbols = ["002521"]
df = load_timeseries_data(stock_symbol=stock_symbols)
df = remove_lunch_break_data(df)
df = drop_zero_columns(df)
df = create_features(df)
df_june = df[(df['TradingDay']>=datetime.datetime(2022, 6, 1)) & (df['TradingDay']<=datetime.datetime(2022, 6, 30))]
df_july = df[(df['TradingDay']>=datetime.datetime(2022, 7, 1)) & (df['TradingDay']<=datetime.datetime(2022, 7, 31))]
df_august = df[(df['TradingDay']>=datetime.datetime(2022, 8, 1)) & (df['TradingDay']<=datetime.datetime(2022, 8, 31))]
df_september = df[(df['TradingDay']>=datetime.datetime(2022, 9, 1)) & (df['TradingDay']<=datetime.datetime(2022, 9, 30))]

used_df = df
df_filtered =used_df[(used_df['InstrumentID']==int(stock_symbols[0]))]
```

```python
def interactive_price_volume_plot(selected_day):
    """
    Generate Price and Volume analysis plots for the selected day.
    Filters the main dataframe for rows with `DataTime` on the given day,
    then creates a 7-panel figure (similar to your original cell [3]) showing:
      - Price Flow (AskPrice1 and BidPrice1)
      - Volume Flow (AskVolume1, BidVolume1, Volume)
      - Average Price
      - Volume
      - Last Price
      - Turnover
      - Volume Change
    """
    import matplotlib.pyplot as plt
    import matplotlib.dates as mdates
    
    day_str = str(selected_day)
    # Filter data for the selected day using the date portion of `DataTime`
    day_data = df_filtered[df_filtered['DataTime'].dt.date == selected_day].copy()
    
    if day_data.empty:
        print("No data available for", selected_day)
        return

    # Create the figure with 7 subplots
    fig, axes = plt.subplots(7, 1, figsize=(20, 12))
    
    # Price Flow Analysis Plot: AskPrice1 and BidPrice1
    axes[0].plot(day_data['DataTime'], day_data['AskPrice1'], color='red', label='Ask Price 1')
    axes[0].plot(day_data['DataTime'], day_data['BidPrice1'], color='green', label='Bid Price 1')
    axes[0].set_title(f'Price Flow Analysis ({day_str})')
    axes[0].set_ylabel('Price')
    axes[0].legend()
    axes[0].grid(True)
    axes[0].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    
    # Volume Flow Analysis Plot: AskVolume1, BidVolume1, Volume
    axes[1].plot(day_data['DataTime'], day_data['AskVolume1'], color='red', label='Ask Volume 1')
    axes[1].plot(day_data['DataTime'], day_data['BidVolume1'], color='green', label='Bid Volume 1')
    if 'Volume' in day_data.columns:
        axes[1].plot(day_data['DataTime'], day_data['Volume'], color='blue', linestyle='--', label='Volume')
    axes[1].set_title(f'Volume Flow Analysis ({day_str})')
    axes[1].set_ylabel('Volume')
    axes[1].set_xlabel('Time')
    axes[1].legend()
    axes[1].grid(True)
    axes[1].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    axes[1].set_yscale('log')
    
    # Average Price Plot
    if 'AveragePrice' in day_data.columns:
        axes[2].plot(day_data['DataTime'], day_data['AveragePrice'], color='blue', label='Average Price')
        axes[2].set_title(f'Average Price Analysis ({day_str})')
        axes[2].set_ylabel('Price')
        axes[2].legend()
        axes[2].grid(True)
        axes[2].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    
    # Volume Plot
    if 'Volume' in day_data.columns:
        axes[3].plot(day_data['DataTime'], day_data['Volume'], color='blue', linestyle='--', label='Volume')
        axes[3].set_title(f'Volume Analysis ({day_str})')
        axes[3].set_ylabel('Volume')
        axes[3].set_xlabel('Time')
        axes[3].legend()
        axes[3].grid(True)
        axes[3].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
        axes[3].set_yscale('log')
    
    # Last Price Plot
    if 'LastPrice' in day_data.columns:
        axes[4].plot(day_data['DataTime'], day_data['LastPrice'], color='purple', label='Last Price')
        axes[4].set_title(f'Last Price Analysis ({day_str})')
        axes[4].set_ylabel('Price')
        axes[4].legend()
        axes[4].grid(True)
        axes[4].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    
    # Turnover Plot
    if 'Turnover' in day_data.columns:
        axes[5].plot(day_data['DataTime'], day_data['Turnover'], color='orange', linestyle='--', label='Turnover')
        axes[5].set_title(f'Turnover Analysis ({day_str})')
        axes[5].set_ylabel('Amount')
        axes[5].set_xlabel('Time')
        axes[5].legend()
        axes[5].grid(True)
        axes[5].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
        axes[5].set_yscale('log')
    
    # Volume Change Plot
    if 'VolumeChange' in day_data.columns:
        axes[6].plot(day_data['DataTime'], day_data['VolumeChange'], color='blue', label='Volume Change')
        axes[6].set_title(f'Volume Change Analysis ({day_str})')
        axes[6].set_ylabel('Volume Change')
        axes[6].set_xlabel('Time')
        axes[6].legend()
        axes[6].grid(True)
        axes[6].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    
    plt.tight_layout()
    plt.show()
```

```python
def interactive_return_plot(selected_day):
    """
    Generate Return analysis plots for the selected day.
    The function creates a 4-panel figure for:
      - Return_1min
      - Return_5min
      - Return_10min
      - Return_1h
    """
    import matplotlib.pyplot as plt
    import matplotlib.dates as mdates

    day_str = str(selected_day)
    day_data = df_filtered[df_filtered['DataTime'].dt.date == selected_day].copy()
    
    if day_data.empty:
        print("No data available for", selected_day)
        return
        
    fig, axes = plt.subplots(4, 1, figsize=(12, 30))
    
    if 'Return_1min' in day_data.columns:
        axes[0].plot(day_data['DataTime'], day_data['Return_1min'], color='purple', label='Return')
        axes[0].set_title(f'Return_1min Analysis ({day_str})')
        axes[0].set_ylabel('Return_1min')
        axes[0].set_xlabel('Time')
        axes[0].legend()
        axes[0].grid(True)
        axes[0].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    
    if 'Return_5min' in day_data.columns:
        axes[1].plot(day_data['DataTime'], day_data['Return_5min'], color='purple', label='Return')
        axes[1].set_title(f'Return_5min Analysis ({day_str})')
        axes[1].set_ylabel('Return_5min')
        axes[1].set_xlabel('Time')
        axes[1].legend()
        axes[1].grid(True)
        axes[1].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    
    if 'Return_10min' in day_data.columns:
        axes[2].plot(day_data['DataTime'], day_data['Return_10min'], color='purple', label='Return')
        axes[2].set_title(f'Return_10min Analysis ({day_str})')
        axes[2].set_ylabel('Return_10min')
        axes[2].set_xlabel('Time')
        axes[2].legend()
        axes[2].grid(True)
        axes[2].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    
    if 'Return_1h' in day_data.columns:
        axes[3].plot(day_data['DataTime'], day_data['Return_1h'], color='purple', label='Return')
        axes[3].set_title(f'Return_1h Analysis ({day_str})')
        axes[3].set_ylabel('Return_1h')
        axes[3].set_xlabel('Time')
        axes[3].legend()
        axes[3].grid(True)
        axes[3].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    
    plt.tight_layout()
    plt.show()
```

```python
def interactive_volatility_trade_plot(selected_day):
    """
    Generate analysis plots for volatility, trade frequency, and order book depth using data from the selected day.
    The function re-computes the rolling volatility (5-minute), 30-minute realized volatility,
    the trade count per minute, and the aggregated order book metrics.
    """
    import matplotlib.pyplot as plt
    import matplotlib.dates as mdates
    import numpy as np

    day_str = str(selected_day)
    day_data = df_filtered[df_filtered['DataTime'].dt.date == selected_day].copy()
    day_data = day_data.sort_values('DataTime')
    
    if day_data.empty:
        print("No data available for", selected_day)
        return

    # Compute 5-minute rolling volatility using 1min returns
    day_data['Volatility_5min'] = day_data['Return_1min'].rolling(window=5, min_periods=1).std() * 100
    
    # Compute 30-min realized volatility as the absolute sum of returns in each 30min interval
    try:
        realized_vol = day_data.resample('30min', on='DataTime')['Return_1min'].apply(lambda x: x.abs().sum())
        day_data['RealizedVol_30min'] = realized_vol.reindex(day_data.index, method='ffill')
    except Exception as e:
        day_data['RealizedVol_30min'] = np.nan

    # Compute trade counts per minute (assuming each row is a trade)
    try:
        day_data['Trades_1min'] = day_data.resample('1min', on='DataTime')['Volume'].transform('count')
    except Exception as e:
        day_data['Trades_1min'] = np.nan

    # Compute aggregated order book metrics (sum of ask/bid volumes)
    ask_bid_cols = [f'AskVolume{i}' for i in range(1, 11)] + [f'BidVolume{i}' for i in range(1, 11)]
    if set(ask_bid_cols).issubset(day_data.columns):
        day_data['TotalBookOrders'] = day_data[ask_bid_cols].sum(axis=1)
        try:
            day_data['BookOrders_1min'] = day_data.resample('1min', on='DataTime')['TotalBookOrders'].transform('sum').ffill()
        except Exception as e:
            day_data['BookOrders_1min'] = np.nan

    # Create a continuous plot index for the x-axis
    day_data['plot_index'] = np.arange(len(day_data))
    # Set a few ticks (start, middle, end)
    tick_positions = [0, len(day_data) // 2, len(day_data) - 1]
    tick_labels = [day_data['DataTime'].iloc[0].strftime('%H:%M:%S'),
                   day_data['DataTime'].iloc[len(day_data)//2].strftime('%H:%M:%S'),
                   day_data['DataTime'].iloc[-1].strftime('%H:%M:%S')]
    
    fig, axes = plt.subplots(3, 1, figsize=(20, 15))
    
    # Volatility Plot
    axes[0].plot(day_data['plot_index'], day_data['Volatility_5min'], color='blue', label='5-min Volatility')
    axes[0].plot(day_data['plot_index'], day_data['RealizedVol_30min'], color='red', linestyle='--', label='30-min Realized Vol')
    axes[0].set_title(f'Volatility Analysis ({day_str})')
    axes[0].legend()
    axes[0].grid(True)
    axes[0].set_xticks(tick_positions)
    axes[0].set_xticklabels(tick_labels, rotation=45)
    
    # Trade Count Plot
    axes[1].plot(day_data['plot_index'], day_data['Trades_1min'], color='green', label='Trades per Minute')
    axes[1].set_title(f'Trade Frequency Analysis ({day_str})')
    axes[1].legend()
    axes[1].grid(True)
    axes[1].set_xticks(tick_positions)
    axes[1].set_xticklabels(tick_labels, rotation=45)
    
    # Order Book Depth Plot
    if 'TotalBookOrders' in day_data.columns and 'BookOrders_1min' in day_data.columns:
        axes[2].plot(day_data['plot_index'], day_data['TotalBookOrders'], color='purple', label='Instant Book Orders')
        axes[2].plot(day_data['plot_index'], day_data['BookOrders_1min'], color='orange', linestyle='--', label='1-min Aggregate Orders')
        axes[2].set_title(f'Order Book Depth Analysis ({day_str})')
        axes[2].legend()
        axes[2].grid(True)
        axes[2].set_xticks(tick_positions)
        axes[2].set_xticklabels(tick_labels, rotation=45)
    
    plt.tight_layout()
    plt.show()
```

```python
from ipywidgets import interact

# Create a sorted list of unique days from the DataTime column
unique_days = sorted(df_filtered['DataTime'].dt.date.unique())

print("Select a day to update the Price & Volume plots:")
interact(interactive_price_volume_plot, selected_day=unique_days)

print("Select a day to update the Return plots:")
interact(interactive_return_plot, selected_day=unique_days)

print("Select a day to update the Volatility, Trade Frequency and Order Book plots:")
interact(interactive_volatility_trade_plot, selected_day=unique_days)
```
